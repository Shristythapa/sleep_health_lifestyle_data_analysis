```python
path = "D:\Sleep_health_and_lifestyle_dataset.csv"
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

```

```python
df = pd.DataFrame(pd.read_csv(path))
```

```python
# Descriptive Statistics for quantitative columns
quantitative_columns = ['Age', 'Sleep Duration', 'Heart Rate', 'Daily Steps']
descriptive_stats = df[quantitative_columns].describe()
descriptive_stats
```

```python
# Distribution of categorical variables
categorical_columns = ['Gender', 'Occupation', 'BMI Category', 'Sleep Disorder']

# Count plot for categorical variables
plt.figure(figsize=(15, 10))

for i, column in enumerate(categorical_columns,1):
    plt.subplot(4, 1, i)
    sns.countplot(data=df, x=column)
    plt.title(f'Distribution of {column}')

plt.tight_layout()
plt.show()
```

```python


age_values_male = df[df['Gender'] == 'Male']['Age'].tolist()
age_values_female = df[df['Gender'] == 'Female']['Age'].tolist()


total_male_count = len(age_values_male)
total_female_count = len(age_values_female)


plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.hist(age_values_male, bins=range(min(age_values_male), max(age_values_male) + 1, 1), edgecolor='black', alpha=0.7, label='Male')
plt.title('Male Age Distribution\nTotal Male Count: {}'.format(total_male_count))
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(age_values_female, bins=range(min(age_values_female), max(age_values_female) + 1, 1), edgecolor='black', alpha=0.7, color='orange', label='Female')
plt.title('Female Age Distribution\nTotal Female Count: {}'.format(total_female_count))
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

```

```python
sleep_duration = df["Sleep Duration"].tolist()
quality_of_sleep=df["Quality of Sleep"].tolist()

# create correlation matrix
matrix = np.corrcoef(sleep_duration, quality_of_sleep)
 
# print matrix
print(matrix)
```

```python
correlation_matrix=df[['Age','Sleep Duration','Quality of Sleep','Physical Activity Level','Stress Level','Heart Rate','Daily Steps']].corr()
```

```python
# Plotting the heatmap
plt.figure(figsize=(5, 5))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=.1)
plt.title('Correlation Matrix Heatmap')
plt.show()
```

High correlation notable
1. Age and quality of sleep
2. Quality of sleep and sleep duration
3. Heart rate and stress level
4. Age and physical activity
5. Quality of sleep and physical activity level

Low correlation notable
1. Stress level and quality of sleep
2. Age and stress level

```python
plot_variables = ['Age','Quality of Sleep','Physical Activity Level','Stress Level','Heart Rate']

sns.pairplot(df[plot_variables], diag_kind='kde',hue='Age',palette="crest")
```

```python

plt.scatter(df[df['Gender']=='Male']['Physical Activity Level'],df[df['Gender']=='Male']['Quality of Sleep'],color='b',label='Male',alpha=0.5)
plt.scatter(df[df['Gender']=='Female']['Physical Activity Level'],df[df['Gender']=='Female']['Quality of Sleep'],color='pink',label='Female',alpha=0.5)

# Add labels and title
plt.xlabel('Physical Activity Level')
plt.ylabel('Quality of Sleep')
plt.title('Scatter Plot')
plt.legend()

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Define a dictionary for colors
category_colors = {'Male': 'b', 'Female': 'pink'}

# Loop through categories and plot
for category, color in category_colors.items():
    subset = df[df['Gender'] == category]
    plt.scatter(subset['Physical Activity Level'], subset['Quality of Sleep'], color=color, label=category, alpha=0.5)

# Add labels and title
plt.xlabel('Physical Activity Level')
plt.ylabel('Quality of Sleep')
plt.title('Scatter Plot')
plt.legend()

# Show the plot
plt.show()
```
