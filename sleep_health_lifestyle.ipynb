```python


gender_counts = df['Gender'].value_counts()


plt.figure(figsize=(8, 6))
gender_counts.plot(kind='bar', color=['blue', 'pink']) 
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.xticks(rotation=0) 
plt.show()
```

```python

age_counts = df["Age"].value_counts()

plt.figure(figsize=(10, 6))
plt.hist(age_counts.index, bins=len(age_counts), weights=age_counts, edgecolor='black', alpha=0.7)
plt.title('Age Value Count Histogram')
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python


age_values_male = df[df['Gender'] == 'Male']['Age'].tolist()
age_values_female = df[df['Gender'] == 'Female']['Age'].tolist()


total_male_count = len(age_values_male)
total_female_count = len(age_values_female)


plt.figure(figsize=(14, 6))

plt.subplot(1, 2, 1)
plt.hist(age_values_male, bins=range(min(age_values_male), max(age_values_male) + 1, 1), edgecolor='black', alpha=0.7, label='Male')
plt.title('Male Age Distribution\nTotal Male Count: {}'.format(total_male_count))
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(True)
plt.legend()

plt.subplot(1, 2, 2)
plt.hist(age_values_female, bins=range(min(age_values_female), max(age_values_female) + 1, 1), edgecolor='black', alpha=0.7, color='orange', label='Female')
plt.title('Female Age Distribution\nTotal Female Count: {}'.format(total_female_count))
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(True)
plt.legend()

plt.tight_layout()
plt.show()

```

```python
sleep_duration = df["Sleep Duration"].tolist()
quality_of_sleep=df["Quality of Sleep"].tolist()

# create correlation matrix
matrix = np.corrcoef(sleep_duration, quality_of_sleep)
 
# print matrix
print(matrix)
```
